```python
import pandas as pd
import numpy as np
df = pd.read_excel('./Harris Data/Expo_Purchasing_join.xlsx')

# print(df)

df.drop(columns=['Late Performance'])
print(df.dtypes)

print(list(df))

df['newCritical'] = np.nan
pd.to_numeric(df['newCritical'], downcast='float')
# print(df.dtypes)
# print(df['newCritical'])
for i, row in df.iterrows():
    if df['Critical'][i]:
        df.set_value(i,'newCritical',1.0)
    else:
        df.set_value(i,'newCritical',.5)

df.drop(columns = ['Critical'])

print(df['newCritical'])

df['newLateNeed'] = np.nan
pd.to_numeric(df['newLateNeed'], downcast='float')
# print(df.dtypes)
# print(df['newCritical'])
for i, row in df.iterrows():
    if df['Late Need'][i]:
        df.set_value(i,'newLateNeed',1.0)
    else:
        df.set_value(i,'newLateNeed',.5)

df.drop(columns = ['Late Need'])

print(df['newLateNeed'])
```

```python
# pd.to_numeric(df['PurchLeadTime'], downcast='float')
df['PurchLeadTime'] = df['PurchLeadTime'].astype(float)
# print(df.dtypes)
df.drop('Critical', axis=1, inplace=True)
df.drop('Late Need', axis=1, inplace=True)
df.drop('Late Performance', axis=1, inplace=True)
print(df.dtypes)
print(df['PurchLeadTime'])
```

```python
from sklearn import preprocessing

y = df.filter(items=['Vendor Code', 'Part Number','ABC','newCritical', 'newLateNeed'])
x = df.filter(items=['StdCost', 'PoCost', 'Qty', 'InternalCostSavings', 'Spend', 'PurchLeadTime'])
x
print(x.dtypes)
x2 = x.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x2)
x3=pd.DataFrame(x_scaled, columns=x.columns)

df2= pd.concat([y, x3], axis=1)
df2
```

```python
print(list(df2))
```

```python
df2['RiskSum'] = np.nan
pd.to_numeric(df2['RiskSum'], downcast='float')
# print(df2.dtypes)
# print(df2['newCritical'])
for i, row in df2.iterrows():
    k = (df2['ABC'][i]+df2['Spend'][i])*(df2['newCritical'][i]+df2['newLateNeed'][i]
                                         +df2['StdCost'][i]+df2['PoCost'][i]+df2['Qty'][i]
                                         +df2['InternalCostSavings'][i])
    df2.set_value(i,'RiskSum',k)
df2
```
