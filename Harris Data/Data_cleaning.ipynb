```python
import pandas as pd
ws = pd.read_excel('ExpoArchive(2).xlsx')
```

```python
from datetime import datetime as dt
def convert_to_date(s):
    try: 
        return dt.date(dt.strptime(str(s)[:10], '%Y-%M-%d'))
    except (TypeError, ValueError) as e:
        return (pd.NaT)
```

```python
ws['Confirmed Dock Date'] = ws['Confirmed Dock Date'].apply(convert_to_date)
ws['Performance Date'] = ws['Performance Date'].apply(convert_to_date)
ws['Need Date'] = ws['Need Date'].apply(convert_to_date)
ws['Order Date'] = ws['Order Date'].apply(convert_to_date)
del ws['Late']
```

```python
print len(ws)
ws = ws.dropna()
print len(ws)
```

```python
ws['temp_Late'] =  ws['Performance Date'] - ws['Confirmed Dock Date']
ws['temp_Late'] = ws['temp_Late'].astype('timedelta64[D]')

ws['Late'] = [i if i > 0 else 0 for i in ws['temp_Late']]
ws['Early'] = [abs(i) if i < 0 else 0 for i in ws['temp_Late']]

del ws['temp_Late']


ws
```

```python
ws.to_excel('ExpoArchive_cleaned.xlsx', index = False)
```
