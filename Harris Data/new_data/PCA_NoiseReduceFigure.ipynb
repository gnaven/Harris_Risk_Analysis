```python
#Plotting an example of PCA noise reduction on a 2d plane.
```

```python
#Import
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
#Generate data
rng = np.random.RandomState(1)
plt.figure(figsize=(20,12))
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal');
```

```python
#PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
#Draw Principal Components 1 and 2
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=2,
                    shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# plot data
plt.figure(figsize=(20,12))
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal');
```

```python
# plot data
pca = PCA(n_components=2)
pca.fit(X)
plt.figure(figsize=(20,12))
plt.scatter(X[:, 0], X[:, 1], alpha=0.3, color = 'black')
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal');

pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)
X_new = pca.inverse_transform(X_pca)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8, color = 'cyan')
plt.axis('equal');

import matplotlib.patches as mpatches
a = mpatches.Patch(color='grey', label='Original')
b = mpatches.Patch(color='cyan', label='Reduced')
leg = plt.legend(handles =[a,b],labels = ['Original','Reduced'],prop={'size': 30},frameon=True,bbox_to_anchor=(.25, 1))

plt.xlabel('Dimension1', fontsize = 30)
plt.ylabel('Dimension2', fontsize = 30)
plt.title('PCA removal of component 2 from 2D data',fontsize = 40)
plt.savefig("PCA_NoiseExample")
plt.show()
```
