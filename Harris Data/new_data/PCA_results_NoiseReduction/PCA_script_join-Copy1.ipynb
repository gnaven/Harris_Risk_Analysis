```python
#Import data
import pandas as pd
DATAFILE = "./prepared_data/join_binary_scale.csv"
Purchase_DF = pd.read_csv(DATAFILE)
print(list(Purchase_DF))
Purchase_DF
```

```python
## Initialization

#Select features
# PCA_feats = ['mean_unexpcost', 'mean_return_spend', 'mean_qty', 'mean_costsave', 'mean_spend',
#              'mean_leadtime','mean_abc', 'mean_source',
#              'mean_late', 'mean_early','%Critical', 'quality_risk_mean', 'delivery_risk_mean']
PCA_feats = ['mean_unexpcost', 'mean_return_spend','mean_spend',
             'mean_late', 'mean_early','%Critical', 'quality_risk_mean', 'delivery_risk_mean']
feats = PCA_feats
#init colnames for later
colnames = ['comp1','comp2','comp3','comp4','comp5','comp6']

Purch_PCA = Purchase_DF[PCA_feats]
Purch_PCA = Purch_PCA.dropna()
# Purch_PCA
```

```python
#Imports
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np
from sklearn.decomposition import FactorAnalysis
```

```python
#Varimax Algorithm
def varimax(Phi, gamma = 1.0, q = 20, tol = 1e-6):
    from numpy import eye, asarray, dot, sum, diag
    from numpy.linalg import svd
    p,k = Phi.shape
    R = eye(k)
    d=0
    for i in range(q):
        d_old = d
        Lambda = dot(Phi, R)
        u,s,vh = svd(dot(Phi.T,asarray(Lambda)**3 - (gamma/p) * dot(Lambda, diag(diag(dot(Lambda.T,Lambda))))))
        R = dot(u,vh)
        d = sum(s)
        if d/d_old < tol: break
    return dot(Phi, R)
```

```python
#RiskSum Algorithm
def RiskSum(df):
    df['RiskSum'] = np.nan
    pd.to_numeric(df['RiskSum'], downcast='float')
    for i, row in df.iterrows():
        k = 0
        for f in feats:
            k = k + df[f][i]
        df['RiskSum'][i] = k
    return df
```

```python
# Run Risk Sum
df1 = pd.read_csv(DATAFILE)
df1 = RiskSum(df1)
df1
```

```python
#Run PCA and generate columns from fit_transform
X = Purch_PCA
pca = PCA(n_components = 6)
x = pca.fit_transform(X)
df_pca = pd.DataFrame(data = x, columns = colnames)

dfc = df1.copy()
df_pca = pd.concat([dfc, df_pca], axis = 1, sort=False)
df_pca
```

```python
#Run FA and generate columns from fit_transform
X = Purch_PCA
fa = FactorAnalysis(n_components = 6)
x = fa.fit_transform(X)
df_fa = pd.DataFrame(data = x, columns = colnames)

dfc = df1.copy()
df_fa = pd.concat([dfc, df_fa], axis = 1, sort=False)
df_fa
```

```python
#Run PCA + varimax and generate columns from fit_transform

X = Purch_PCA
pca = PCA(n_components = 6)
pca.fit(X)
vm = varimax(pca.components_)
pca.components_ = vm
x = pca.transform(X)
df_vm = pd.DataFrame(data = x, columns = colnames)

dfc = df1.copy()
df_vm = pd.concat([dfc, df_vm], axis = 1, sort=False)
df_vm
```

```python
df_pca.to_csv("PCA_reduced.csv", index = False)
df_vm.to_csv("PCA_reduced_varimax.csv", index = False)
df_fa.to_csv("FA_reduced.csv", index = False)
```
