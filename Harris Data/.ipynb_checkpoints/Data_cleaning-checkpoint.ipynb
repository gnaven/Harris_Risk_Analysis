```python
import pandas as pd
ws = pd.read_excel('ExpoArchive(2).xlsx')
```

```python
ws
```

```python
from datetime import datetime as dt
def convert_to_date(s):
    try: 
        return dt.date(dt.strptime(str(s)[:10], '%Y-%M-%d'))
    except (TypeError, ValueError) as e:
        return (pd.NaT)
```

```python
ws['Confirmed Dock Date'] = ws['Confirmed Dock Date'].apply(convert_to_date)
ws['Performance Date'] = ws['Performance Date'].apply(convert_to_date)
ws['Need Date'] = ws['Need Date'].apply(convert_to_date)
ws['Order Date'] = ws['Order Date'].apply(convert_to_date)
del ws['Late']
```

```python
ws = ws.dropna()
len(ws)
```

```python
ws['Late Need'] =  ws['Confirmed Dock Date'] - ws['Need Date']
ws['Late Performance'] =  ws['Confirmed Dock Date'] - ws['Performance Date']
```

```python
ws['Late Need'] = ws['Late Need'].astype('timedelta64[D]')
ws['Late Performance'] = ws['Late Performance'].astype('timedelta64[D]')

ws
```
