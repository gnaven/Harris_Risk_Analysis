```python
import pandas as pd

join = pd.read_excel('category_aggregate/join_category.xlsx')
#ptool = pd.read_excel('category_aggregate/ptool_category.xlsx')
```

```python
import matplotlib.pyplot as plt
import numpy as np

#join = join.drop('Category', axis = 1)
join = join.drop('var_log_stdcost', 'var_log_pocost', 'var_log_qty', 'var_log_return_qty', 'var_log_costsave', 
                 'var_log_spend', 'var_log_return_spend', 'variance_log_leadtime')

for feat in join.columns:
    plt.hist(join[feat], bins = 50)
    plt.title(feat)
    plt.xlabel = ("Value")
    plt.ylabel = ("Count")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

features = ['median_stdcost', 'median_pocost', 'median_qty',
       'median_costsave', 'sum_costsave', 'median_spend', 'sum_spend', 'median_leadtime', 'mean_priority',
           '%_singlesource', 'avg_late', 'avg_early', '%_late', '%_early', '%_critical']

categories = join['Category'].drop_duplicates()

for feat in features:
    count = 1
    f, axs = plt.subplots(3,5,figsize=(25,25))
    for cat in categories:
        current_feat = join[(join['Category'] == cat)][feat]
        plt.subplot(3, 5, count)
        plt.hist(current_feat, bins = 50)
        print "Category: " + cat + ", Feature: " + feat
        #print "std: " + str(np.std(current_feat))
        #print "mean: " + str(np.mean(current_feat))
        #print "median: " + str(np.median(current_feat))
        #print "max: " + str(np.max(current_feat))
        #print "min: " + str(np.min(current_feat)) + "\n\n"
        plt.title(cat + ", " + feat)
        plt.xlabel = ("Value")
        plt.ylabel = ("Count")
        count += 1
    fname = "vendor_category_distr/" + feat + "_distr"
    plt.savefig(fname)
    plt.show()
```
