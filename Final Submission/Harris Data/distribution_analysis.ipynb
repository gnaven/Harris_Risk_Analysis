```python
import pandas as pd

join = pd.read_csv('join.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

features = ['StdCost', 'PoCost', 'Qty',
           'InternalCostSavings', 'Spend', 'PurchLeadTime', 'ABC',  
            'transaction_type', 'Return_spend', 'Return_qty', 'Source',
           'Critical', 'Late',
           'Early']

for feat in features:
    plt.hist(join[feat], bins = 15)
    plt.title(feat)
    plt.xlabel = ("Value")
    plt.ylabel = ("Count")
    fname = "distr/" + feat + "_distr"
    plt.savefig(fname)
    plt.show()
```
